### Manipulação de dados exemplo (3)

````python
# Create a dictionary with student data
student_data = {
    "Name": ["Alex", "Ben", "Clara", "Deric", "Eva", "Frank"],
    "Age": [20, 21, 19, 22, 23, 20],
    "Major": ["CS", "Math", "Physics", "Economics", "Biology", "Chemistry"],
    "Score": [85, 92, 78, 95, 88, 90],
    "Hometown": ["York", "Angeles", "Chicago", "Houston", "Boston", "Francisco"],
    "Graduation Year": [2022, 2023, 2022, 2023, 2024, 2023],
    "Scholarship": [False, True, False, True, False, False]
}

# Create a DataFrame from the dictionary
df3 = pd.DataFrame(student_data)

# Print the DataFrame 
print(df3)
````

````python
df3.describe()
````

````python
# exemplos de filtragem de dados
# df3.["Age"]
# df3[["Age", "Score"]]
````

````python
# filtrando valores condicionais
# df3[df3["Score"] > 89]
df3[(df3["Age"] > 20) | (df3["Score"] > 94)] 
````

### Manipulação de dados exemplo (4)
* Mostrando como funciona o LOC e ILOC
* ```df.loc[<linhas>, <colunas>]```


````python
df4 = pd.read_csv("https://raw.githubusercontent.com/dataPalacio/datasets/refs/heads/main/nba-draft-2015/nba_draft_2015.csv")
df4.head()
````

````python
df4.shape
````

````python
df4.info()
````

````python
# retornando 10 linhas
df4.loc[0:10]
````

````python
# visualizando 3 linhas com mais de 1 condicional
df_nba = df4.set_index('Player')
df_nba.loc[["Devin Booker", "Trey Lyles", "Justise Winslow"], ["Draft Year", "Superstar"]]	
````

````python
# setando o valor 0.15 para os 20 primeiros
df4.loc[0:20, "Superstar"] = 0.15
df4.head(25)
````

````python
# trabalhando com ILOC
#df4.iloc[0:5] # selecione a primeira até a quarta linha
#df4.iloc[0:10,0] # selecione da primeira até a décima linha da primeira coluna
df4.iloc[0:10, [0,1,5]] # selecione as dez primeiras linhas e apenas a primeira, segunda e quinta coluna
````
